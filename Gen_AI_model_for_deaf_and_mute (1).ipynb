```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("grassknoted/asl-alphabet")

print("Path to dataset files:", path)
```

```python
import tensorflow as tf

dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "/kaggle/input/asl-alphabet/asl_alphabet_train/asl_alphabet_train", # Corrected nested path
    image_size=(64,64),
    batch_size=32
)

print(f"Found {len(dataset.file_paths)} files belonging to {len(dataset.class_names)} classes.")
```

```python
for images,labels in dataset.take(1):
  print(images.shape)
```

##normalization

```python
norm= tf.keras.layers.Rescaling(1./255)
dataset= dataset.map(lambda x,y:(norm(x),y))
```

##train and validation

```python
train_size=int(0.8* len(dataset))
train_dataset=dataset.take(train_size)
val_dataset=dataset.skip(train_size)
```

##CNN model

```python
model= tf.keras.Sequential([
    tf.keras.layers.Conv2D(32,(3,3), activation='relu', input_shape=(64,64,3)),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(64,(3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(128,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(29, activation='softmax')
])
```

##compile model

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
history= model.fit(
    train_dataset,
    validation_data= val_dataset,
    epochs=10,
)
```

```python
test_acc, test_loss= model.evaluate(val_dataset)
```
